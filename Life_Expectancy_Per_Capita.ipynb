## Standarization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=life_df.drop(columns=['District'])
scaler.fit(X)
scaled_data=scaler.transform(X)
scaled_data.shape
```

## Clustering

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
n_neighbors=5
neigh= NearestNeighbors(n_neighbors=n_neighbors)
nbrs= neigh.fit(scaled_data)
distances, indices = nbrs.kneighbors(scaled_data)
distances.shape
```

```python
distances =np.sort(distances, axis =0)
distances[0]
distances_=distances[:,n_neighbors-1]
plt.plot(distances_)
plt.grid(color='grey', linestyle='-',axis='y',alpha=0.2)
plt.yticks(np.arange(0,3,0.25))
plt.show()
```

```python
dbscan=DBSCAN(eps=0.34, min_samples=3)
dbscan.fit(scaled_data)
life_df['DBSCAN Label'] = dbscan.labels_
```

```python
dbscan.labels_
```

```python
pd.set_option("Max_rows",100)
life_df.sort_values(by='DBSCAN Label')
```
